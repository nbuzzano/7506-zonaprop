#### 5) inflacion inmobiliaria en Mexico.

```python
%matplotlib inline

import math
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot  as plt

from datetime import datetime

pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
# Creamos el dataframe y le asignamos como id del dataframe, el id de cada row del csv.
df = pd.read_csv('train_transformed.csv', index_col='id')
```

```python
df5 = df.copy()

df5 = df5[['precio','fecha']]
df5.dropna(inplace= True)

df5['fecha'] = df5['fecha'].apply(func = (lambda x :(datetime.strptime(x, '%Y-%m-%d %H:%M:%S'))))
```

```python
df_by_months = df5['fecha'].apply(func = (lambda t : (t.year-2012)*12+t.month))
df5 = df5.groupby(df_by_months).precio.mean()
df5.plot(c='r',linewidth =2.5, figsize = [18,8])

plt.title('Inflacion inmobiliaria Mexico')
plt.xticks(np.arange(0,61,12),np.arange(2012,2018,1))
plt.xlabel('AÃ±o')
plt.ylabel('Mean price\n (pesos)')
plt.show()
```
